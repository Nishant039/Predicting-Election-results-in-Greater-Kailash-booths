```python
df_13_15_bjp_aap=pd.DataFrame()
df_13_15_bjp_aap["bjp_13"]=df_13.iloc[:,0]
df_13_15_bjp_aap["aap_13"]=df_13.iloc[:,4]
df_13_15_bjp_aap["bjp_15"]=df_15.iloc[:,1]
df_13_15_bjp_aap["aap_15"]=df_15.iloc[:,4]


df_13_15_bjp_aap.head()
```

```python
df_13_15_bjp_aap["bjp13-aap13"]=df_13_15_bjp_aap["bjp_13"]-df_13_15_bjp_aap["aap_13"]
df_13_15_bjp_aap["bjp15-aap15"]=df_13_15_bjp_aap["bjp_15"]-df_13_15_bjp_aap["aap_15"]

df_13_15_bjp_aap["% change in difference(15-13)"]=(df_13_15_bjp_aap["bjp15-aap15"]-df_13_15_bjp_aap["bjp13-aap13"])*100/abs(df_13_15_bjp_aap["bjp13-aap13"])
```

```python
df_13_15_bjp_aap.head()
```

```python
df_13_15_bjp_aap["% change in difference(15-13)"].describe()
```

```python
### % Change in difference >+100 means that the difference now has reversed (earlier bjp was lacking but now bjp is in lead ) or bjp was already in lead and now in much more lead.
```

```python
indices=np.where(df_13_15_bjp_aap["% change in difference(15-13)"]>100)
print("Booths-", indices[0]+1)
```
